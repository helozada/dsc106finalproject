```python
for i in range(len(df)):
    if isinstance(df['Performer'][i], list) == True:
        continue
    else:
        df['Performer'][i] = [df['Performer'][i]]
df
```

```python
artists = []
for i in range(len(df)):
    for j in range(len(df['Performer'][i])):
        if df['Performer'][i][j] in artists:
            continue
        else:
            artists.append(df['Performer'][i][j])
artists
```

```python
for artist in artists:
    bools = []
    for i in range(len(df)):
        if artist in df['Performer'][i]:
            bools.append(True)
        else:
            bools.append(False)
    df['In'] = bools
    a = df[df['In'] == True]
    a = a.drop_duplicates(subset=['Song']).groupby('WeekID')['In'].sum()
    df2[artist] = a
df2
```
